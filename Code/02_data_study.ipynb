Aqui calculamos el tama√±o en base a los tokens o en base a los caracteres

```python
#caracteres
#df["size"]=df['tweet'].apply(lambda x: len(x))
#df["size"].describe()

#tokens
df["tokens"]=df['tweet'].apply(lambda x: x.split())
df["size"]=df['tokens'].apply(lambda x: len(x))
df["size"].describe()
```

```python
import seaborn as sns
# matplotlib histogram

plt.hist(df['size'], color = 'lightblue', edgecolor = 'black') # if you do not define bins, plt will obtain the best value for it.

# Add labels
#plt.title('Length distribution of the texts')
plt.xlabel('Length of texts (tokens)')
plt.ylabel('Instances')
```

```python
labels = df["task1"].unique()
for name in labels:
    # Subset to the language
    subset = df[df['task1'] == name]
    if name == "YES":
      value = "Sexist"
    else:
      value = "Not Sexist"
    # Draw the density plot
    sns.distplot(subset['size'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = value)

# Plot formatting
plt.legend(prop={'size': 10}, title = 'Type')
#plt.title('Density Plot for sexist and non-sexist posts (task1)')
plt.xlabel('Length of the text (characters)')
plt.ylabel('Density')
```

```python
df4 = df2[df2["task2"]!='UNKNOWN']

labels = df4["task2"].unique()
for name in labels:
    # Subset to the language
    subset = df4[df4['task2'] == name]

    # Draw the density plot
    sns.distplot(subset['size'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = name)

# Plot formatting
plt.legend(prop={'size': 10}, title = 'Categories')
# plt.title('Density Plot for sexist posts (task2)')
plt.xlabel('Length of the text (characters)')
plt.ylabel('Density')
```
